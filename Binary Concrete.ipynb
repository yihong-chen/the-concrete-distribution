# The Binary Concrete Distribution
The binary concrete distribution is defined on $(0, 1)$. Let's see how it looks! The probability density of a binary concrete variable is given as follows.

![Binary Concrete](https://upload-images.jianshu.io/upload_images/5225017-b5a3265630d64038.png?imageMogr2/auto-orient/strip%7CimageView2/2/w/1240)

```python
def binary_concrete(alpha, beta):
    """
    
    args:
        logalpha: location
        beta: temperature
    """
    def binary_concrete_parametrized(s):
        frac1 = beta * alpha * (s ** (- beta - 1)) * ((1 - s) ** (- beta - 1)) 
        frac2 = (alpha * (s ** (- beta)) + ((1 - s) ** (- beta))) ** 2
        return frac1 / frac2
    return binary_concrete_parametrized
```

Set the location parameter $\alpha = 1$, which should approximate Bernoulli($\frac{\alpha}{\alpha + 1} = 0.5$),
and the temperature parameter $\beta = 0.5$.

```python
f = binary_concrete(alpha=1, beta=0.5)
```

Let's see its distribution over $(0,1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.arange(0.01, 1, 0.01)
z_density = f(z)
plt.plot(z, z_density)
plt.xlabel('z')
plt.ylabel('p(z)')
plt.show()
```

How will it change if we use different temperatures? For example, $\beta = 10$

```python
f = binary_concrete(alpha=1, beta=10)

z = np.arange(0.01, 1, 0.01)
z_density = f(z)
```

```python
plt.plot(z, z_density)
plt.xlabel('z')
plt.ylabel('p(z)')
plt.show()
```

$\alpha=1$, $\beta=10$ produces something concentrate on $0.5$. Let's try $\beta=1$.

```python
f = binary_concrete(alpha=1, beta=1)
z = np.arange(0.01, 1, 0.01)
z_density = f(z)
plt.plot(z, z_density)
plt.xlabel('z')
plt.ylabel('p(z)')
plt.show()
```

What about $\alpha=1$, $\beta=0.9$?

```python
f = binary_concrete(alpha=1, beta=0.9)
z = np.arange(0.01, 1, 0.01)
z_density = f(z)
plt.plot(z, z_density)
plt.xlabel('z')
plt.ylabel('p(z)')
plt.show()
```

And $\alpha=1, \beta=0.01$

```python
f = binary_concrete(alpha=1, beta=0.01)
z = np.arange(0.01, 1, 0.01)
z_density = f(z)
plt.plot(z, z_density)
plt.xlabel('z')
plt.ylabel('p(z)')
plt.show()
```

To sum up, for $\alpha=1$, which means if we want to approximate _Bernoulli(0.5)_ , choosing $\beta \in (0,1)$ seems promising. For different $\alpha$s, choosing $\beta \in (0,1)$ concentrates the probability mass near the two endpoints.

```python
def save_binary_concrete_fig(alpha, beta):
    f = binary_concrete(alpha, beta)
    z = np.arange(0.01, 1, 0.01)
    z_density = f(z)
    plt.plot(z, z_density)
    plt.xlabel('z')
    plt.ylabel('p(z)')
    plt.title('alpha={},beta={}'.format(alpha, beta))
    plt.savefig('figs/alpha{}_beta{}.png'.format(alpha, beta))
    plt.show()
```

```python
alpha = 4
for beta in [20, 10, 5, 2, 1, 0.8, 0.4, 0.2, 0.1, 0.01]:
    save_binary_concrete_fig(alpha, beta)
```
